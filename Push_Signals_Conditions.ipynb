```python
import livef1
import pandas as pd
from seeq import spy
import scipy
```

```python
#year_list = [2024,2025]
current_season = 2025

#put a workbook name in the line below if you would like the signals scoped to a workbook else use None to make the signals global
workbook_name = None
datasource_name = 'F1 Telemetry Playground'
```

```python
car_number_info = pd.read_csv("CarTeamMap.csv")
car_number_info
```

```python
season = livef1.get_season(current_season)
#pd.set_option('display.max_rows', None)
```

```python
# Ensure session data is parsed
season.parse_sessions()

# Access the DataFrame containing session details
sessions_df = season.season_table  
sessions_df['session_startDate'] = pd.to_datetime(sessions_df['session_startDate'])
sessions_df['gmtoffset'] = pd.to_timedelta(sessions_df['gmtoffset'])


#sessions_df = sessions_df[sessions_df.session_key>9993]
#sessions_df = sessions_df[sessions_df.session_key<9993]
sessions_df['Capsule Start'] = sessions_df['session_startDate'] + sessions_df['gmtoffset']
sessions_df['Capsule End'] = sessions_df['session_endDate'] + sessions_df['gmtoffset']
```

```python
session_df['start']
```

```python
for index, row in sessions_df.iterrows():

    current_meeting = index
    #current_meeting = 1252
    current_session = row['session_key']
    #current_session = 9662

    

    
    print(str(current_session))
    
    session = livef1.get_session(season = current_season,
                             meeting_key=current_meeting,
                             session_key = current_session
                            )
    try:
        session.generate(silver = True)
    except Exception as e:
        print(f"Skipping session {current_session} due to error: {e}")
        continue  # Move on to the next row
        
    laps_data = session.get_laps()
    telemetry_data = session.get_car_telemetry()

    telemetry_data = telemetry_data.dropna(subset = ['DriverNo'])
    telemetry_data['DriverNo'] = pd.to_numeric(telemetry_data['DriverNo'], errors='coerce').astype('Int64')
    telemetry_data = telemetry_data.drop(columns=[ 'SessionKey'])

    laps_data = laps_data.dropna(subset = ['DriverNo'])
    laps_data['DriverNo'] = pd.to_numeric(laps_data['DriverNo'], errors='coerce').astype('Int64')

    cars_in_session = telemetry_data.DriverNo.unique()

    for current_DriverNo in cars_in_session:
        print(str(current_DriverNo))

        matching_row = car_number_info[(car_number_info['Season'] == current_season) & (car_number_info['DriverNo'] == current_DriverNo)]

        # Extract the TeamName value
        team_name = matching_row['TeamName'].values
        driver_name = matching_row['DriverName'].values
        
        # Return the first match if multiple values exist
        current_team_name = team_name[0] if len(team_name) > 0 else None  
        current_driver_name = driver_name[0] if len(driver_name) > 0 else None

        print(current_team_name)
        print(current_driver_name)


        DriverNo_telemetry_data = telemetry_data[telemetry_data.DriverNo == current_DriverNo]
        DriverNo_telemetry_data = DriverNo_telemetry_data.drop(columns = ['DriverNo','timestamp'])
        DriverNo_telemetry_data['Utc'] = pd.to_datetime(DriverNo_telemetry_data['Utc'])
        DriverNo_telemetry_data = DriverNo_telemetry_data.set_index('Utc')
        
        new_column_names = [f"{current_season}.{current_team_name}.{current_DriverNo}.{col}" for col in DriverNo_telemetry_data.columns]
        DriverNo_telemetry_data.columns = new_column_names

        signal_push_results = spy.push(data=DriverNo_telemetry_data, workbook=workbook_name, datasource=datasource_name)

        
        
        DriverNo_laps_data = laps_data[laps_data.DriverNo == current_DriverNo]
        DriverNo_laps_data = DriverNo_laps_data.dropna(subset = ['lap_time'])
        
        DriverNo_laps_data ['team_name'] = current_team_name
        DriverNo_laps_data ['current_driver_name'] = current_driver_name

        sector_1_caps = DriverNo_laps_data
        sector_1_caps['Capsule Start'] = sector_1_caps['lap_start_date']
        sector_1_caps['Capsule End'] = sector_1_caps['Capsule Start'] + sector_1_caps['sector1_time']
        sector_1_caps['Sector'] = 1
        
        sector_1_caps = sector_1_caps[['lap_number','in_pit','DriverNo','Capsule Start','Capsule End','Sector','team_name','current_driver_name']]
        
        sector_1_caps = sector_1_caps.dropna(subset = ['Capsule Start'])
        sector_1_caps = sector_1_caps.dropna(subset = ['Capsule End'])

        sector_2_caps = DriverNo_laps_data
        sector_2_caps['Capsule Start'] = sector_2_caps['lap_start_date']+sector_2_caps['sector1_time']
        sector_2_caps['Capsule End'] = sector_2_caps['Capsule Start']+sector_2_caps['sector2_time']
        sector_2_caps['Sector'] = 2
        
        sector_2_caps = sector_2_caps[['lap_number','in_pit','DriverNo','Capsule Start','Capsule End','Sector','team_name','current_driver_name']]
        
        sector_2_caps = sector_2_caps.dropna(subset = ['Capsule Start'])
        sector_2_caps = sector_2_caps.dropna(subset = ['Capsule End'])

        sector_3_caps = DriverNo_laps_data
        sector_3_caps['Capsule Start'] = sector_3_caps['lap_start_date']+sector_3_caps['sector1_time']+sector_3_caps['sector2_time']
        sector_3_caps['Capsule End'] = sector_3_caps['Capsule Start']+sector_3_caps['sector3_time']
        sector_3_caps['Sector'] = 3
        
        sector_3_caps = sector_3_caps[['lap_number','in_pit','DriverNo','Capsule Start','Capsule End','Sector','team_name','current_driver_name']]
        
        sector_3_caps = sector_3_caps.dropna(subset = ['Capsule Start'])
        sector_3_caps = sector_3_caps.dropna(subset = ['Capsule End'])

        all_sectors = pd.concat([sector_1_caps, sector_2_caps, sector_3_caps], ignore_index=True)

        spy.push(data=all_sectors,
             metadata=pd.DataFrame([{
                 'Name': f"{current_season}.{current_team_name}.{current_DriverNo}.Sectors",
                 'Type': 'Condition',
                 'Maximum Duration': '5 min',
                 
             }]), workbook=workbook_name, datasource=datasource_name)



        

        

        

        

    

    
```
